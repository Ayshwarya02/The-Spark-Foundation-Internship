```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dataset =  pd.read_csv("SampleSuperstore.csv")
```

```python
dataset.head(10)
```

```python
dataset.describe()
```

```python
dataset.nunique(axis=1)
```

```python
dataset.nunique(axis=0)
```

```python
corr = dataset.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, 
            cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

```python
sns.catplot(x="Segment",y="Profit",data=dataset,kind='box',ci=95)
```

```python
sns.catplot(x="Ship Mode",y="Profit",data=dataset,kind='bar',ci=None)
```

```python
order = dataset.groupby(["State"])["Profit"].mean().sort_values().index
sns.catplot(y="State",x="Profit",data=dataset,kind='bar',ci=None,height=20,order=order)
```

```python
sns.catplot(x="Category",y="Profit",data=dataset,kind="bar",ci=None)
```

```python
sns.catplot(x="Category",y="Profit",data=dataset,kind='bar',ci=None,col="Region")
```

```python
plt.subplots(figsize=(8,10))
order=dataset.groupby(['State'])["State"].count().sort_values().index
sns.countplot(y='State',data=dataset,order=order)
```
